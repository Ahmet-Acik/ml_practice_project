```python
def generate_scenario_data(scenario_name, scenario_config, random_state=42):
    """
    Generate diverse student data for a specific educational scenario
    
    This function creates realistic student performance data based on
    the unique characteristics of different educational environments.
    """
    np.random.seed(random_state)
    n_samples = scenario_config['target_samples']
    
    # Extract configuration
    base_score = scenario_config['base_exam_score']
    study_range = scenario_config['study_hours_range']
    attend_range = scenario_config['attendance_range']
    grade_range = scenario_config['previous_grade_range']
    sleep_range = scenario_config['sleep_hours_range']
    noise = scenario_config['noise_level']
    weights = scenario_config['factor_weights']
    
    print(f"🎓 Generating {n_samples:,} samples for {scenario_name.replace('_', ' ').title()}...")
    
    # Generate base features with scenario-specific distributions
    study_hours = np.random.uniform(study_range[0], study_range[1], n_samples)
    attendance = np.random.uniform(attend_range[0], attend_range[1], n_samples)
    previous_grade = np.random.uniform(grade_range[0], grade_range[1], n_samples)
    sleep_hours = np.random.uniform(sleep_range[0], sleep_range[1], n_samples)
    
    # Scenario-specific feature variations
    if scenario_name == 'elite_private':
        # Elite schools: more tutoring, higher family support
        family_support = np.random.uniform(7, 10, n_samples)
        extra_activities = np.random.uniform(2, 8, n_samples)  # Many activities
        
    elif scenario_name == 'urban_public':
        # Urban public: diverse family support, moderate activities
        family_support = np.random.uniform(3, 9, n_samples)
        extra_activities = np.random.uniform(0, 6, n_samples)
        
    elif scenario_name == 'rural_community':
        # Rural: strong family values, fewer activities
        family_support = np.random.uniform(5, 10, n_samples)
        extra_activities = np.random.uniform(0, 4, n_samples)
        
    elif scenario_name == 'stem_magnet':
        # STEM magnet: academic-focused families, intense activities
        family_support = np.random.uniform(6, 10, n_samples)
        extra_activities = np.random.uniform(3, 7, n_samples)
        
    elif scenario_name == 'arts_creative':
        # Arts schools: supportive but flexible families
        family_support = np.random.uniform(4, 9, n_samples)
        extra_activities = np.random.uniform(1, 5, n_samples)
        
    else:  # international
        # International: varied family support, cultural activities
        family_support = np.random.uniform(4, 9, n_samples)
        extra_activities = np.random.uniform(1, 6, n_samples)
    
    # Calculate exam scores using scenario-specific weights
    exam_score = (
        base_score +
        weights['study_hours'] * (study_hours - study_range[0]) / (study_range[1] - study_range[0]) * 15 +
        weights['attendance'] * (attendance - attend_range[0]) / (attend_range[1] - attend_range[0]) * 10 +
        weights['previous_grade'] * (previous_grade - grade_range[0]) / (grade_range[1] - grade_range[0]) * 20 +
        weights['sleep_hours'] * (8 - np.abs(sleep_hours - 7.5)) * 2 +
        weights['family_support'] * family_support * 1.5 +
        weights['extra_activities'] * extra_activities * 0.8 +
        np.random.normal(0, noise * 10, n_samples)  # Scenario-specific noise
    )
    
    # Ensure realistic bounds
    exam_score = np.clip(exam_score, 0, 100)
    
    # Create DataFrame
    scenario_data = pd.DataFrame({
        'study_hours': study_hours,
        'attendance': attendance,
        'previous_grade': previous_grade,
        'sleep_hours': sleep_hours,
        'family_support': family_support,
        'extra_activities': extra_activities,
        'exam_score': exam_score,
        'scenario': scenario_name
    })
    
    print(f"   ✅ Generated {len(scenario_data):,} samples")
    print(f"   📊 Exam Score Range: {scenario_data['exam_score'].min():.1f} - {scenario_data['exam_score'].max():.1f}")
    print(f"   📈 Mean Score: {scenario_data['exam_score'].mean():.1f}")
    
    return scenario_data

print("🔧 Data generation function ready!")
```

```python
# Phase 3A: Advanced Feature Engineering
print("🔧 PHASE 3A: ADVANCED FEATURE ENGINEERING")
print("="*50)
print("🎯 Goal: Create sophisticated features from diverse educational data")
print("📊 Input: 4,600+ samples across 6 educational scenarios")
print()

# Prepare feature matrix (exclude target and scenario)
X_diverse = diverse_educational_data.drop(['exam_score', 'scenario'], axis=1)
y_diverse = diverse_educational_data['exam_score']
scenarios = diverse_educational_data['scenario']

print(f"📈 Base Features: {X_diverse.shape[1]}")
print(f"📊 Samples: {X_diverse.shape[0]:,}")
print()

# 1. Create interaction features
print("🔗 Creating Interaction Features...")
from sklearn.preprocessing import PolynomialFeatures

# Create polynomial features (degree 2 for interactions)
poly_features = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_poly = poly_features.fit_transform(X_diverse)

# Get feature names
feature_names = poly_features.get_feature_names_out(X_diverse.columns)
X_poly_df = pd.DataFrame(X_poly, columns=feature_names)

print(f"   ✅ Generated {X_poly_df.shape[1]} features (including interactions)")
print(f"   📊 New interaction features: {X_poly_df.shape[1] - X_diverse.shape[1]}")
print()

# 2. Create scenario-based features
print("🏫 Creating Scenario-Based Features...")
from sklearn.preprocessing import LabelEncoder

# Encode scenarios as numerical features
le = LabelEncoder()
scenario_encoded = le.fit_transform(scenarios)

# Add scenario features
X_enhanced = X_poly_df.copy()
X_enhanced['scenario_encoded'] = scenario_encoded

# Create scenario-specific feature interactions
for i, col in enumerate(X_diverse.columns):
    X_enhanced[f'{col}_scenario_interaction'] = X_enhanced[col] * X_enhanced['scenario_encoded']

print(f"   ✅ Added scenario encoding and interactions")
print(f"   📊 Total features: {X_enhanced.shape[1]}")
print()

# 3. Feature scaling and selection preparation
print("📏 Preparing Feature Scaling...")
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_enhanced)
X_scaled_df = pd.DataFrame(X_scaled, columns=X_enhanced.columns)

print(f"   ✅ Scaled {X_scaled_df.shape[1]} features")
print(f"   📊 Feature matrix shape: {X_scaled_df.shape}")
print()

# 4. Feature importance analysis
print("🎯 Analyzing Feature Importance...")

# Select top features using F-statistic
k_best = 25  # Select top 25 features
selector = SelectKBest(score_func=f_regression, k=k_best)
X_selected = selector.fit_transform(X_scaled, y_diverse)

# Get selected feature names
selected_features = X_enhanced.columns[selector.get_support()]
X_final = pd.DataFrame(X_selected, columns=selected_features)

print(f"   ✅ Selected top {k_best} features from {X_enhanced.shape[1]} total")
print(f"   🎯 Final feature matrix: {X_final.shape}")
print()

print("🔥 TOP SELECTED FEATURES:")
feature_scores = selector.scores_[selector.get_support()]
feature_importance = pd.DataFrame({
    'feature': selected_features,
    'score': feature_scores
}).sort_values('score', ascending=False)

for i, (_, row) in enumerate(feature_importance.head(10).iterrows()):
    print(f"   {i+1:2d}. {row['feature']:<30} (score: {row['score']:.1f})")

print()
print("✨ FEATURE ENGINEERING COMPLETE!")
print(f"📊 Ready for advanced modeling with {X_final.shape[1]} optimized features")
```
